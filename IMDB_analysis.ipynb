```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('imdb_top_1000.csv')  
print(df.head())  
print(df.info()) 
```

```python
# Удаляем ненужные колонки
df = df.drop(columns=['Poster_Link', 'Overview'])
```

```python
# Проверяем пропуски
print(df.isnull().sum())
```

```python
# Заполним пропуски средним значением
df['Meta_score'] = df['Meta_score'].fillna(df['Meta_score'].mean())
```

```python
# Проверяем дубликаты по колонкам 'Series_Title' и 'Released_Year'
partial_duplicates = df.duplicated(subset=['Series_Title', 'Released_Year']).sum()
print(f"Частичных дубликатов (по названию и году): {partial_duplicates}")
```

```python
# Проверяем статистику числовых колонок
print(df.describe())
```

```python
# Смотрим на распределение рейтингов
plt.figure(figsize=(10, 6))
sns.histplot(df['IMDB_Rating'], bins=20, kde=True)
plt.title('Распределение рейтингов IMDb')
plt.xlabel('Рейтинг')
plt.ylabel('Количество фильмов')
plt.show()
```

```python
# Топ-10 жанров
df['Genre'] = df['Genre'].str.split(', ')  
genres = df.explode('Genre')['Genre'].value_counts().head(10)  

plt.figure(figsize=(12, 6))
sns.barplot(x=genres.index, y=genres.values)
plt.title('Топ-10 жанров')
plt.xticks(rotation=45)
plt.show()
```

```python
# Убедимся, что 'Released_Year' - число (если это строка, преобразуем)
if df['Released_Year'].dtype == 'object':
    df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')

# Строим scatter plot + тренд-линию
plt.figure(figsize=(12, 6))
sns.regplot(
    x='Released_Year', 
    y='IMDB_Rating', 
    data=df, 
    scatter_kws={'alpha': 0.4, 'color': 'blue'},  # Прозрачность и цвет точек
    line_kws={'color': 'red', 'linewidth': 2}    # Цвет и толщина линии тренда
)
plt.title('Зависимость рейтинга IMDb от года выпуска фильма', fontsize=14)
plt.xlabel('Год выпуска', fontsize=12)
plt.ylabel('Рейтинг IMDb', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
top_directors = df['Director'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.index, y=top_directors.values)
plt.title('Топ-10 режиссёров по количеству фильмов')
plt.xticks(rotation=45)
plt.show()
```
